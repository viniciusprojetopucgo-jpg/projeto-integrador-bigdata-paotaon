```python
from pyspark.sql import functions as F

# 1. CONEXÃO
driver = "org.postgresql.Driver"
url = "jdbc:postgresql://gondola.proxy.rlwy.net:31447/railway"
table = "public.vendas_pao_ta_on"
user = "postgres"
password = "Sua_senha_aq"

df = spark.read.format("jdbc") \
    .option("driver", driver).option("url", url) \
    .option("dbtable", table).option("user", user) \
    .option("password", password).load()

# 2. Análise por Bairro (Ajustado para total_faturado)
vendas_por_bairro = df.groupBy("bairro").agg(
    F.count("id").alias("quantidade_pedidos"),
    F.sum("valor").alias("total_faturado")
)
print("Análise de Vendas por Bairro:")
display(vendas_por_bairro.orderBy(F.desc("total_faturado")))

# 3. Cruzamento: Bairro x Origem (Sugestão 2)
cruzamento_bairro_origem = df.groupBy("bairro", "origem").agg(
    F.count("id").alias("quantidade_pedidos"),
    F.sum("valor").alias("total_faturado")
)
print("Cruzamento: Preferência de Canal por Bairro:")
display(cruzamento_bairro_origem.orderBy("bairro", F.desc("total_faturado")))

# 4. Volumetria de Itens (A que você pediu para manter)
volumetria_estoque = df.groupBy("produto").agg(
    F.count("produto").alias("unidades_vendidas")
)
print("Volumetria de Itens (Foco em Estoque):")
display(volumetria_estoque.orderBy(F.desc("unidades_vendidas")))
```
